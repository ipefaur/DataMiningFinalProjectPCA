```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
from keras.layers import Dense, Dropout, Activation, Embedding, LSTM, SpatialDropout1D
import tensorflow as tf
```

```python
df = pd.read_csv("mushrooms.csv")
comp = 3 #22 components atm
```

```python
# Convert our code to numbers.
#Thank you at: https://stackoverflow.com/questions/40435350/pandas-convert-text-feature-to-numeric-value
df = df.apply(lambda col : pd.factorize(col, sort=False)[0])
df
```

```python
#pop the label to be trained with later.
label = pd.get_dummies(df.pop("class"))
```

```python
#Scale the dataframe
sc = StandardScaler()
scData = sc.fit_transform(df)
df = pd.DataFrame(scData, index=df.index, columns=df.columns) 
df
```

```python
#Apply PCA
pca = PCA(n_components=comp)

pc = pca.fit_transform(df)
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

```python
df = pd.DataFrame(pc, columns=['PC%s' % (i + 1) for i in range(comp)])
df
```

```python
#MACHINE LEARNING MODEL!
feature_train, feature_test, label_train, label_test = train_test_split(df, label, test_size=0.2, random_state=0)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(comp, input_shape=(comp,), activation="relu"))
model.add(tf.keras.layers.Dense(2,activation="softmax")) # 2 output types (good vs bad)
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss="binary_crossentropy", metrics=["accuracy"])
model.build()
model.summary() #summarize the modfel
```

```python
model.fit(feature_train, label_train, epochs=5, shuffle = True)
```

```python
test_loss, test_acc = model.evaluate(feature_test, label_test) #evaluate the model
print("ACCURACY: ", test_acc)
```
